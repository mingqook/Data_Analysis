## 3. 데이터 살펴보기
### 3.1 데이터프레임의 구조

```python
import pandas as pd
import os

df = pd.read_csv(os.path.join(os.getcwd(), 'auto-mpg.csv'), header = None )
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
```

```python
# head - 데이터프레임의 앞부분 일부 출력 / tail - 데이터프레임의 뒷부분 일부 출력
print(df.head())
print("-------------------------------------------------------------------")
print(df.tail())
```

```python
# shape - 데이터프레임의 행과 열의 개수 출력
print(df.shape)
print("----------------------------------------")

# info() - 데이터프레임에 관한 기본 정보 출력(클래스 유형, 행 인덱스의 구성, 열 이름의 종류와 개수, 열의 자료형과 개수 등)
print(df.info())
print("----------------------------------------")

# dtypes - 데이터프레임의 자료형 확안
print(df.dtypes)
print("----------------------------------------")
```

```python
# describe() - 데이터프레임의 열에 대한 주요 기술통계정보 요약 후 출력 / include = 'all' 추가 시 산술데이터가 아닌 열에 대한 정보 포함
print(df.describe())
print("----------------------------------------")
print(df.describe(include = 'all'))
```

```python
# count() - 데이터프레임의 각 열이 가지고 있는 데이터 개수를 시리즈 객체로 반환 / 유효한 값의 개수만 계산
print(df.count())
print("----------------------------------------")

# value_counts() - 시리즈 객체의 고유값 개수를 세는 데 사용 / dropna = True 사용 시 nan 제외
print(df['origin'].value_counts())
print("----------------------------------------")
```

### 3.2 통계 함수 적용

```python
# mean() - 산술 데이터를 갖는 모든 열의 평균값을 계산
print(df.mean())
print("----------------------------------------")
print('mpg의 평균:', df['mpg'].mean())
print("----------------------------------------")

# median() - 산술 데이터를 갖는 모든 열의 중앙값을 계산
print(df.median())
print("----------------------------------------")
print('mpg의 중앙값:', df['mpg'].median())
print("----------------------------------------")

# max() - 산술 데이터를 갖는 모든 열의 최대값을 계산
print(df.max())
print("----------------------------------------")
print('mpg의 최대값:', df['mpg'].max())
print("----------------------------------------")

# min() - 산술 데이터를 갖는 모든 열의 최소값을 계산
print(df.min())
print("----------------------------------------")
print('mpg의 최소값:', df['mpg'].min())
print("----------------------------------------")

# std() - 산술데이터를 갖는 모든 열의 표준편차를 계산
print(df.std())
print("----------------------------------------")
print('mpg의 표준편차:', df['mpg'].std())
print("----------------------------------------")

# corr() - 산술데이터를 갖는 2개의 열의 상관계수를 계산
print(df.corr())
print("----------------------------------------")
```

### 3.3 판다스 내장 그래프 도구 활용
#### - 판다스는 matplotlib 라이브러리 기능을 일부 내장하고 있어서, 별도로 임포트하지 않고도 간단한 그래프를 그릴 수 있다.

```python
# excel file load를 위해 패키지 설치
# !pip install openpyxl
```

```python
df = pd.read_excel(os.path.join(os.getcwd(), '남북한발전전력량.xlsx'))
df_ns = df.iloc[[0,5], 3:]
df_ns.index = ['south','north']
df_ns.columns = df_ns.columns.map(int)
print(df_ns.head())
```

```python
# plot() - 추가 옵션을 설정하지 않으면 선 그래프 
print(df_ns.plot())
print("----------------------------------------")

# 열 이름을 구성하고 있는 연도 값이 행 인덱스에 위치하도록 행렬을 전치해줘야 함
tdf_ns = df_ns.T
print(tdf_ns.plot())
print("----------------------------------------")
```

```python
# plot(kind = 'bar') - 막대 그래프 
tdf_ns = df_ns.T
print(tdf_ns.plot(kind = 'bar'))
print("----------------------------------------")
```

```python
# plot(kind = 'hist') - 히스토그램
tdf_ns = df_ns.T
print(tdf_ns.plot(kind = 'hist'))
print("----------------------------------------")
```
