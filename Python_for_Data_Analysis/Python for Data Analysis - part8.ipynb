### 8.1 계층적 색인

```python
import pandas as pd
import numpy as np

data = pd.Series(np.random.randn(9), index = [['a','a','a','b','b','c','c','d','d'],
                                             [1,2,3,1,3,1,2,2,3]])

print(data)
print('--------------------------------------')

print(data.index)
print('--------------------------------------')

print(data['b'])
print('--------------------------------------')

print(data['b':'c'])
print('--------------------------------------')

print(data.loc[['b','d']])
print('--------------------------------------')

print(data.loc[:,2])
print('--------------------------------------')

# unstack 메서드 - data를 새롭게 배열
print(data.unstack())
print('--------------------------------------')

# stack 메서드 - unstack의 반대
print(data.unstack().stack())
print('--------------------------------------')
```

```python
frame = pd.DataFrame(np.arange(12).reshape(4,3), index = [['a','a','b','b'], [1,2,1,2]], 
                    columns = [['ohio', 'ohio', 'colorado'], ['green', 'red', 'green']])

print(frame)
print('--------------------------------------')

frame.index.names = ['key1', 'key2']
frame.columns.names = ['state', 'color']
print(frame)
print('--------------------------------------')

print(frame['ohio'])
print('--------------------------------------')
```

#### 8.1.1 계층의 순서를 바꾸고 정렬하기

```python
# swaplevel 메서드 - 계층 번호나 이름이 뒤바뀐 새로운 객체를 반환(데이터는 변경되지 않음)
print(frame.swaplevel('key1', 'key2'))
print('--------------------------------------')

# sort_index 메서드 - 단일 계층에 속한 데이터를 정렬
print(frame.sort_index(level = 1))
print('--------------------------------------')

print(frame.swaplevel(0,1).sort_index(level = 0))
print('--------------------------------------')
```

#### 8.1.2 계층별 요약 통계

```python
print(frame.sum(level = 'key2'))
print('--------------------------------------')

print(frame.sum(level = 'color', axis = 1))
print('--------------------------------------')
```

#### 8.1.3 DataFrame의 컬럼 사용하기

```python
frame = pd.DataFrame({'a' : range(7), 'b' : range(7,0,-1), 
                     'c' : ['one', 'one', 'one', 'two', 'two', 'two', 'two'], 
                     'd' : [0,1,2,0,1,2,3]})

print(frame)
print('--------------------------------------')

# set_index - 하나 이상의 컬럼을 색인으로 하는 새로운 DataFrame 생성
frame2 = frame.set_index(['c', 'd'])
print(frame2)
print('--------------------------------------')

print(frame.set_index(['c', 'd'], drop = False))
print('--------------------------------------')

# reset_index - 계층적 색인 단계가 컬럼으로 이동
print(frame2.reset_index())
print('--------------------------------------')
```

### 8.2 데이터 합치기
#### - pandas.merge : 하나 이상의 키를 기준으로 DataFrame의 로우를 합침
#### - pandas.concat : 하나의 축을 따라 객체를 이어 붙임
#### - combile_first : 두 객체를 포개서 한 객체에서 누락된 데이터를 다른 객체에 있는 값으로 채움

#### 8.2.1 데이터베이스 스타일로 DataFrame 합치기

```python
# pd.merge() - 하나 이상의 키를 사용해서 데이터 집합의 로우를 합침
df1 = pd.DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1' : range(7)})
df2 = pd.DataFrame({'key' : ['a', 'b', 'd'], 
                   'data2' : range(3)})

print(df1)
print('--------------------------------------')
print(df2)
print('--------------------------------------')

print(pd.merge(df1, df2))
print('--------------------------------------')

print(pd.merge(df1, df2, on = 'key'))
print('--------------------------------------')

df3 = pd.DataFrame({'lkey' : ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 
                   'data1' : range(7)})
df4 = pd.DataFrame({'rkey' : ['a','b', 'd'], 
                   'data2' : range(3)})

print(pd.merge(df3, df4, left_on = 'lkey', right_on='rkey'))
print('--------------------------------------')

# 외부조인 - 합집합 / 왼쪽조인, 오른쪽조인 - 왼쪽, 오른쪽의 모든 로우를 포함
print(pd.merge(df1, df2, how ='outer'))
print('--------------------------------------')
```

#### how 옵션에 따른 조인 연산 - inner(공통으로 존재하는 키), left(왼쪽 테이블에 존재하는 모든 키), right(오른쪽 테이블에 존재하는 모든 키), output(양쪽에 존재하는 모든 키)

```python
# 다대다 조인은 두 로우의 데카르트 곱을 반환
df1 = pd.DataFrame({'key' : ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1' : range(6)})
df2 = pd.DataFrame({'key' : ['a', 'b', 'a', 'b', 'd'], 
                   'data2' : range(5)})

print(df1)
print('--------------------------------------')

print(df2)
print('--------------------------------------')

print(pd.merge(df1, df2, on='key', how = 'left'))
print('--------------------------------------')

print(pd.merge(df1, df2, how = 'inner'))
print('--------------------------------------')
```

```python
left = pd.DataFrame({'key1' : ['foo', 'foo', 'bar'], 
                    'key2' : ['one', 'two', 'one'], 
                    'lval' : [1,2,3]})
right = pd.DataFrame({'key1' : ['foo', 'foo', 'bar', 'bar'], 
                     'key2' : ['one', 'one', 'one', 'two'], 
                     'rval' : [4,5,6,7]})

print(pd.merge(left, right, on = ['key1', 'key2'], how = 'outer'))
print('--------------------------------------')

print(pd.merge(left, right, on = 'key1'))
print('--------------------------------------')

print(pd.merge(left, right, on = 'key1', suffixes=('_left', '_right')))
print('--------------------------------------')
```

#### merge 함수 인자 - left(왼족 df), right(오른쪽 df), how(조인 방법, inner, outer, left, right), on(조인하려는 칼럼 이름, 양쪽 모두 존재해야 함), left_on(조인 키로 사용할 왼쪽 df 칼럼), right_on(조인 키로 사용할 오른쪽 df 칼럼), left_index(조인 키로 사용할 왼쪽 df의 색인), right_index(조인 키로 사용할 오른쪽 df의 색인), sort(조인키에 따라 병합된 데이터를 정렬), suffixes(컬럼 이름이 겹칠 경우 이름 뒤에 붙일 문자열)

#### 8.2.2 색인 병합하기

```python
left1 = pd.DataFrame({'key' : ['a','b','a','a','b','c'], 
                     'value' : range(6)})
right1 = pd.DataFrame({'group_val' : [3.5, 7]},
                      index = ['a', 'b'])

print(left1)
print('--------------------------------------')

print(right1)
print('--------------------------------------')

print(pd.merge(left1, right1, left_on = 'key', right_index = True))
print('--------------------------------------')

print(pd.merge(left1, right1, left_on = 'key', right_index = True, how = 'outer'))
print('--------------------------------------')
```

```python
lefth = pd.DataFrame({'key1' : ['ohio', 'ohio', 'ohio', 'nevada', 'nevada'], 
                     'key2' : [2000, 2001, 2002, 2001, 2002], 
                     'data' : np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape(6,2), 
                     index = [['nevada', 'nevada','ohio', 'ohio', 'ohio', 'ohio'], 
                             [2001, 2000, 2000, 2000, 2001, 2002]], 
                     columns = ['event1', 'event2'])
print(lefth)
print('--------------------------------------')

print(righth)
print('--------------------------------------')

print(pd.merge(lefth, righth, left_on = ['key1', 'key2'], right_index = True))
print('--------------------------------------')

print(pd.merge(lefth, righth, left_on = ['key1', 'key2'], right_index = True, how = 'outer'))
print('--------------------------------------')
```

```python
left2 = pd.DataFrame([[1,2], [3,4], [5,6]], 
                    index = ['a','c','e'], 
                    columns = ['ohio', 'nevada'])
right2 = pd.DataFrame([[7,8], [9,10], [11, 12], [13, 14]], 
                     index = ['b', 'c', 'd', 'e'], 
                     columns = ['missouri', 'alabama'])

print(left2)
print('--------------------------------------')

print(right2)
print('--------------------------------------')

print(pd.merge(left2, right2, how = 'outer', left_index = True, right_index = True))
print('--------------------------------------')

# join - index 병합에 편리
print(left2.join(right2, how = 'outer'))
print('--------------------------------------')

print(left1.join(right1, on = 'key'))
print('--------------------------------------')
```

```python
another = pd.DataFrame([[7,8], [9,10], [11,12], [16,17]], 
                      index = ['a', 'c', 'e', 'f'], 
                      columns = ['new york', 'oregon'])
print(another)
print('--------------------------------------')

print(left2.join([right2, another]))
print('--------------------------------------')

print(left2.join([right2, another], how = 'outer'))
print('--------------------------------------')
```

#### 8.2.4 겹치는 데이터 합치기

```python
# np.where - 벡터화된 if-else 구문
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan], 
             index = ['f', 'e', 'd','c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64), 
             index = ['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan

print(a)
print('--------------------------------------')

print(b)
print('--------------------------------------')

print(np.where(pd.isnull(a), b, a))
print('--------------------------------------')

# combine_fist - 바로 위 결과와 동일, 데이터 정렬 기능까지 제공
print(b[:-2].combine_first(a[2:]))
```

```python
df1 = pd.DataFrame({'a' : [1, np.nan, 5, np.nan], 
                   'b' : [np.nan, 2, np.nan, 6], 
                   'c' : range(2, 18, 4)})
df2 = pd.DataFrame({'a' : [5,4, np.nan, 3, 7], 
                   'b' : [np.nan, 3, 4, 6, 8]})

print(df1)
print('--------------------------------------')

print(df2)
print('--------------------------------------')

print(df1.combine_first(df2))
print('--------------------------------------')
```

### 8.3 재형성과 피벗
#### - stack : 데이터의 컬럼을 로우로 피벗(회전)시킨다.
#### - unstack : 로우를 컬럼으로 피벗시킨다

```python
data = pd.DataFrame(np.arange(6).reshape(2,3), 
                   index = pd.Index(['ohio', 'colorado'], name = 'state'), 
                   columns = pd.Index(['one', 'two', 'three'], name = 'number'))
print(data)
print('--------------------------------------')

result = data.stack()
print(result)
print('--------------------------------------')

print(result.unstack())
print('--------------------------------------')

print(result.unstack(0))
print('--------------------------------------')

print(result.unstack('state'))
print('--------------------------------------')
```

```python
s1 = pd.Series([0,1,2,3], index = ['a', 'b', 'c', 'd'])
s2 = pd.Series([4,5,6], index = ['c', 'd', 'e'])
data2 = pd.concat([s1,s2], keys = ['one', 'two'])

print(data2)
print('--------------------------------------')

print(data2.unstack())
print('--------------------------------------')

print(data2.unstack().stack())
print('--------------------------------------')

print(data2.unstack().stack(dropna=False))
print('--------------------------------------')
```

```python
df = pd.DataFrame({'left' : result, 'right' : result +5}, 
                 columns=pd.Index(['left', 'right'], name = 'side'))
print(df)
print('--------------------------------------')

print(df.unstack())
print('--------------------------------------')

print(df.unstack('state').stack('side'))
print('--------------------------------------')
```

#### 8.3.2 긴 형식에서 넓은 형식으로 피벗하기

```python
data = pd.read_csv('examples/macrodata.csv')
print(data.head())
print('--------------------------------------')

periods = pd.PeriodIndex(year = data.year, quarter = data.quarter, name = 'date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name = 'item')
data = data.reindex(columns = columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns = {0:'value'})

print(ldata[:10])
print('--------------------------------------')

pivoted = ldata.pivot('date', 'item', 'value')
print(pivoted)
print('--------------------------------------')
```

#### 8.3.3 넓은 형식에서 긴 형식으로 피벗하기

```python
# pd.melt - pivot과 반대되는 연산
df = pd.DataFrame({'key' : ['foo', 'bar', 'baz'], 
                  'A' : [1,2,3],
                  'B' : [4,5,6], 
                  'C' : [7,8,9]})
print(df)
print('--------------------------------------')

melted = pd.melt(df, ['key'])
print(melted)
print('--------------------------------------')

reshaped = melted.pivot('key', 'variable', 'value')
print(reshaped)
print('--------------------------------------')

print(reshaped.reset_index())
print('--------------------------------------')

print(pd.melt(df, id_vars = ['key'], value_vars = ['A', 'B']))
print('--------------------------------------')

print(pd.melt(df, value_vars = ['A', 'B', 'C']))
print('--------------------------------------')
```
